```python
import torch
import torch.nn as nn
import torch.utils.data as data_utils
import pandas as pd
import io
import matplotlib.pyplot as plt
import numpy as np
```

```python
# A simple 2 layer neural network
class NNModel(nn.Module):
  def __init__(self, input_dim, hidden_dim, output_dim):
    super(NNModel, self).__init__()

    self.fc1 = nn.Linear(input_dim, hidden_dim)
    self.sigmoid = nn.Sigmoid()
    self.fc2 = nn.Linear(hidden_dim, output_dim)
  
  def forward(self, x):
    out = self.fc1(x)
    out = self.sigmoid(out)
    out = self.fc2(out)
    return out

loss = nn.MSELoss()
```

```python
def load_tensors(features, labels):
  features_tensor = torch.Tensor(features)
  labels_tensor = torch.Tensor(labels)
  return data_utils.TensorDataset(features_tensor, labels_tensor)

def train_loader(features, labels, batch_size):
  tensor_dataset = load_tensors(features, labels)
  loader = data_utils.DataLoader(tensor_dataset, batch_size=batch_size, shuffle=True)
  return loader
```

```python
#Load binary encoded data
X_binary_test = np.load('../../training_sets/Xtest_binary_set_data.npy', allow_pickle=True)
X_binary_train = np.load('../../training_sets/Xtrain_binary_set_data.npy', allow_pickle=True)
X_binary_val = np.load('../../training_sets/Xvalidate_binary_set_data.npy', allow_pickle=True)

Y_binary_test = np.load('../../training_sets/Ytest_binary_set_data.npy', allow_pickle=True)
Y_binary_train = np.load('../../training_sets/Ytrain_binary_set_data.npy', allow_pickle=True)
Y_binary_val = np.load('../../training_sets/Yvalidate_binary_set_data.npy', allow_pickle=True)
```

```python
def train_model(model, df_features=X_binary_train, df_labels=Y_binary_train, n_iters=10000, batch_size=50, learning_rate=0.01):
    dataframe_len=len(df_features)
    loader = train_loader(features=df_features, labels=df_labels, batch_size=batch_size)
    optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
    num_epochs = int(n_iters / (dataframe_len / batch_size))
    running_losses = []
    running_accuracy = []
    running_epoch = []
    
    for epoch in range(num_epochs):
        for i,(x_train,y_train) in enumerate(loader):
            y_pred = model(x_train)
            cost = loss(y_pred,y_train)
            optimizer.zero_grad()
            cost.backward()
            optimizer.step()
        running_losses.append(cost)
        running_epoch.append(epoch)
        if epoch % 500 == 0 and epoch != 0:
                print('epoch: ' + str(epoch) + ' cost: ' + str(cost))
    with torch.no_grad():
        plt.subplot(122)
        plt.plot(running_epoch, running_losses)
        plt.title('Running Losses vs Epoch')
        plt.show()
    return model
  
```

```python
#binary Encoding
print("First Model")
first_model = NNModel(input_dim=312, hidden_dim=10, output_dim=1)
first_model = train_model(model=first_model, n_iters=50000, batch_size=50, learning_rate=1e-5)
print("Second Model")
second_model = NNModel(input_dim=312, hidden_dim=10, output_dim=1)
second_model = train_model(model=second_model, n_iters=70000, batch_size=50, learning_rate=1.1e-5)
print("Third Model")
third_model = NNModel(input_dim=312, hidden_dim=10, output_dim=1)
third_model = train_model(model=third_model, n_iters=70000, batch_size=50, learning_rate=0.9e-5)
print("Fourth Model")
fourth_model = NNModel(input_dim=312, hidden_dim=10, output_dim=1)
fourth_model = train_model(model=fourth_model, n_iters=50000, batch_size=100, learning_rate=0.8e-5)
print("Fifth Model")
fifth_model = NNModel(input_dim=312, hidden_dim=10, output_dim=1)
fifth_model = train_model(model=fifth_model, n_iters=50000, batch_size=50, learning_rate=0.7e-5)
```

```python
# Let's use the validation sets on our models
tensor_binary_X_val = torch.Tensor(X_binary_val)
tensor_binary_Y_val = torch.Tensor(Y_binary_val)

y_pred = first_model(tensor_binary_X_val)
cost = loss(y_pred, tensor_binary_Y_val)
print("Cost first model: " + str(cost))

y_pred = second_model(tensor_binary_X_val)
cost = loss(y_pred, tensor_binary_Y_val)
print("Cost second model: " + str(cost))

y_pred = third_model(tensor_binary_X_val)
cost = loss(y_pred, tensor_binary_Y_val)
print("Cost third model: " + str(cost))

y_pred = fourth_model(tensor_binary_X_val)
cost = loss(y_pred, tensor_binary_Y_val)
print("Cost fourth model: " + str(cost))

y_pred = fifth_model(tensor_binary_X_val)
cost = loss(y_pred, tensor_binary_Y_val)
print("Cost fifth model: " + str(cost))
```

```python
# Testing on our fourth model
tensor_binary_X_test = torch.Tensor(X_binary_test)
tensor_binary_Y_test = torch.Tensor(Y_binary_test)

y_pred = fourth_model(tensor_binary_X_test)
cost = loss(y_pred, tensor_binary_Y_test)
print("Test fourth model: " + str(cost))
```

```python
class DNNModel(nn.Module):
    def __init__(self, input_dim=312, output_dim=1):
        super(DNNModel, self).__init__()
        
        self.sigmoid = nn.Sigmoid()
        self.relu = nn.ReLU()
        inter_layer_size1 = int((input_dim + output_dim) / 2)
        self.fc1 = nn.Linear(input_dim, inter_layer_size1)
        inter_layer_size2 = int((inter_layer_size1 + output_dim) / 2)
        self.fc2 = nn.Linear(inter_layer_size1, inter_layer_size2)
        inter_layer_size3 = int((inter_layer_size2 + output_dim) / 2)
        self.fc3 = nn.Linear(inter_layer_size2, inter_layer_size3)
        self.fc_last = nn.Linear(inter_layer_size3, output_dim)
    
    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        
        out = self.fc2(out)
        out = self.relu(out)
        
        out = self.fc3(out)
        out = self.relu(out)
        
        out = self.fc_last(out)
        return out
```

```python
#binary Encoding
print("First Model")
first_model = DNNModel(input_dim=312, output_dim=1)
first_model = train_model(model=first_model, n_iters=50000, batch_size=50, learning_rate=1e-5)
print("Second Model")
second_model = DNNModel(input_dim=312, output_dim=1)
second_model = train_model(model=second_model, n_iters=10000, batch_size=50, learning_rate=0.9e-5)
print("Third Model")
third_model = DNNModel(input_dim=312, output_dim=1)
third_model = train_model(model=third_model, n_iters=30000, batch_size=50, learning_rate=0.8e-5)
print("Fourth Model")
fourth_model = DNNModel(input_dim=312, output_dim=1)
fourth_model = train_model(model=fourth_model, n_iters=50000, batch_size=100, learning_rate=1.1e-5)
print("Fifth Model")
fifth_model = DNNModel(input_dim=312, output_dim=1)
fifth_model = train_model(model=fifth_model, n_iters=50000, batch_size=50, learning_rate=0.8e-5)
```

```python
# Let's use the validation sets on our models
tensor_binary_X_val = torch.Tensor(X_binary_val)
tensor_binary_Y_val = torch.Tensor(Y_binary_val)

y_pred = first_model(tensor_binary_X_val)
cost = loss(y_pred, tensor_binary_Y_val)
print("Cost first model: " + str(cost))

y_pred = second_model(tensor_binary_X_val)
cost = loss(y_pred, tensor_binary_Y_val)
print("Cost second model: " + str(cost))

y_pred = third_model(tensor_binary_X_val)
cost = loss(y_pred, tensor_binary_Y_val)
print("Cost third model: " + str(cost))

y_pred = fourth_model(tensor_binary_X_val)
cost = loss(y_pred, tensor_binary_Y_val)
print("Cost fourth model: " + str(cost))

y_pred = fifth_model(tensor_binary_X_val)
cost = loss(y_pred, tensor_binary_Y_val)
print("Cost fifth model: " + str(cost))
```

```python
# Testing on our fourth model
y_pred = fifth_model(tensor_binary_X_test)
cost = loss(y_pred, tensor_binary_Y_test)
print("Test fifth model: " + str(cost))
```

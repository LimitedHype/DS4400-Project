```python
import torch
import torch.nn as nn
import torch.utils.data as data_utils
import pandas as pd
import io
import matplotlib.pyplot as plt

test_df = pd.read_csv('hw04_data/test.csv')
train_df = pd.read_csv('hw04_data/train.csv')
validation_df = pd.read_csv('hw04_data/validation.csv')
```

```python
# split the dataframes

def get_features_labels(df):
  df_features = df.iloc[:, df.columns != 'genre']
  df_labels = df['genre']
  return df_features, df_labels
```

```python
test_df_features, test_df_labels = get_features_labels(test_df)
validation_df_features, validation_df_labels = get_features_labels(validation_df)
train_df_features, train_df_labels = get_features_labels(train_df)

#print unique classes
print("Test set: " + str(test_df_labels.unique()))
print("Train set: " + str(train_df_labels.unique()))
print("Validation set: " + str(validation_df_labels.unique()))

#print number for each class
print("Test set: " + str(test_df_labels.value_counts()))
print("Train set: " + str(train_df_labels.value_counts()))
print("Validation set: " + str(validation_df_labels.value_counts()))
```

```python
# Question 2

def load_tensors(features, labels):
  features = features.values
  labels=pd.get_dummies(labels).values
  features_tensor = torch.Tensor(features)
  labels_tensor = torch.Tensor(labels)
  return data_utils.TensorDataset(features_tensor, labels_tensor)

def train_loader(features, labels, batch_size):
  tensor_dataset = load_tensors(features, labels)
  loader = data_utils.DataLoader(tensor_dataset, batch_size=batch_size, shuffle=True)
  return loader
```

```python
# A simple 2 layer neural network
class NNModel(nn.Module):
  def __init__(self, input_dim, hidden_dim, output_dim):
    super(NNModel, self).__init__()

    self.fc1 = nn.Linear(input_dim, hidden_dim)
    self.sigmoid = nn.Sigmoid()
    self.fc2 = nn.Linear(hidden_dim, output_dim)
  
  def forward(self, x):
    out = self.fc1(x)
    out = self.sigmoid(out)
    out = self.fc2(out)
    return out

loss = nn.CrossEntropyLoss()
```

```python
# Part C

def train_model(model, df_features=train_df_features, df_labels=train_df_labels, n_iters=70000, batch_size=50, learning_rate=0.01, dataframe_=train_df):
  dataframe_len=len(dataframe_)
  loader = train_loader(features=df_features, labels=df_labels, batch_size=batch_size)
  optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
  num_epochs = int(n_iters / (dataframe_len / batch_size))
  running_losses = []
  running_accuracy = []
  running_epoch = []
  for epoch in range(num_epochs):
    total = 0
    correct = 0
    
    class_acc = [0, 0, 0, 0, 0, 0]
    class_total = [0, 0, 0, 0, 0, 0]
    for i,(x_train,y_train) in enumerate(loader):
      y_pred = model(x_train)
      cost = loss(y_pred,y_train)
      _, predicted = torch.max(y_pred.data, 1)
      _, y_train = torch.max(y_train.data, 1)

      total += y_train.size(0)
      correct += (predicted == y_train).sum()
      class_index = predicted[0]
      class_acc[class_index] += (predicted == y_train).sum()
      class_total[class_index] += y_train.size(0)
      optimizer.zero_grad()
      cost.backward()
      optimizer.step()
    running_losses.append(cost)
    running_accuracy.append(correct/total)
    running_epoch.append(epoch)
    if epoch % 50 == 0 and epoch != 0:
      for i, genre in enumerate(train_df_labels.unique()):
        if class_total[i] == 0:
            print(str(genre) + ": " + str(1.0))
        else:
            print(str(genre) + ": " + str((class_acc[i]) / (class_total[i])))
      print('epoch: ' + str(epoch) + ' cost: ' + str(cost))
      print('accuracy: ' + str(correct/total))
  fig, axs = plt.subplots(2)
  fig.suptitle('Losses and accuracy versus epoch')
  axs[0].plot(running_epoch, running_losses)
  axs[1].plot(running_epoch, running_accuracy)
  plt.show()
  return model
```

```python
# Train model with 64 neurons in hidden dim and one hidden layer

print("First Model")
first_model = NNModel(input_dim=300, hidden_dim=64, output_dim=6)
first_model = train_model(model=first_model, n_iters=50000, batch_size=50, learning_rate=0.1)
print("Second Model")
second_model = NNModel(input_dim=300, hidden_dim=64, output_dim=6)
second_model = train_model(model=second_model, n_iters=70000, batch_size=50, learning_rate=0.01)
print("Third Model")
third_model = NNModel(input_dim=300, hidden_dim=64, output_dim=6)
third_model = train_model(model=third_model, n_iters=70000, batch_size=50, learning_rate=0.001)
print("Fourth Model")
fourth_model = NNModel(input_dim=300, hidden_dim=64, output_dim=6)
fourth_model = train_model(model=fourth_model, n_iters=50000, batch_size=100, learning_rate=0.1)
print("Fifth Model")
fifth_model = NNModel(input_dim=300, hidden_dim=64, output_dim=6)
fifth_model = train_model(model=fifth_model, n_iters=50000, batch_size=50, learning_rate=0.01)
```

```python
def other_datasets(model, df_features, df_labels, batch_size=50, dataframe_=validation_df):
  loader = train_loader(features=df_features, labels=df_labels, batch_size=batch_size)
  total = 0
  correct = 0

  for i,(x_train,y_train) in enumerate(loader):
    y_pred = model(x_train)
    cost = loss(y_pred,y_train)
    _, predicted = torch.max(y_pred.data, 1)
    _, y_train = torch.max(y_train.data, 1)
    total += y_train.size(0)
    correct += (predicted == y_train).sum()
  print('Dataset accuracy: ' + str(correct/total))
```

```python
# Train best model/validation
# First Model
print("First model")
other_datasets(model=first_model, 
               df_features=validation_df_features, 
               df_labels=validation_df_labels, 
               dataframe_=validation_df)

# Second Model
print("Second model")
other_datasets(model=second_model, 
               df_features=validation_df_features, 
               df_labels=validation_df_labels, 
               dataframe_=validation_df)

# Third Model
print("Third model")
other_datasets(model=third_model, 
               df_features=validation_df_features, 
               df_labels=validation_df_labels, 
               dataframe_=validation_df)

# Fourth Model
print("Fourth model")
other_datasets(model=fourth_model, 
               df_features=validation_df_features, 
               df_labels=validation_df_labels, 
               dataframe_=validation_df)

# Fifth Model
print("Fifth model")
other_datasets(model=fifth_model, 
               df_features=validation_df_features, 
               df_labels=validation_df_labels, 
               dataframe_=validation_df)
```

```python
# Test best model
other_datasets(model=first_model, 
               df_features=test_df_features, 
               df_labels=test_df_labels, 
               dataframe_=test_df)
```

```python
# Two Hidden Layers
class Two_NNModel(nn.Module):
  def __init__(self, input_dim, hidden_dim, output_dim):
    super(Two_NNModel, self).__init__()

    self.fc1 = nn.Linear(input_dim, hidden_dim)
    self.sigmoid1 = nn.Sigmoid()
    self.fc2 = nn.Linear(hidden_dim, hidden_dim)
    self.sigmoid2 = nn.Sigmoid()
    self.fc3 = nn.Linear(hidden_dim, output_dim)
  
  def forward(self, x):
    out = self.fc1(x)
    out = self.sigmoid1(out)
    out = self.fc2(out)
    out = self.sigmoid2(out)
    out = self.fc3(out)
    return out
```

```python
# Train model with 64 neurons in hidden dim and two hidden layers

print("First Model")
first_model = Two_NNModel(input_dim=300, hidden_dim=64, output_dim=6)
first_model = train_model(model=first_model, n_iters=50000, batch_size=50, learning_rate=0.1)
print("Second Model")
second_model = Two_NNModel(input_dim=300, hidden_dim=64, output_dim=6)
second_model = train_model(model=second_model, n_iters=70000, batch_size=50, learning_rate=0.01)
print("Third Model")
third_model = Two_NNModel(input_dim=300, hidden_dim=64, output_dim=6)
third_model = train_model(model=third_model, n_iters=70000, batch_size=50, learning_rate=0.001)
print("Fourth Model")
fourth_model = Two_NNModel(input_dim=300, hidden_dim=64, output_dim=6)
fourth_model = train_model(model=fourth_model, n_iters=50000, batch_size=100, learning_rate=0.1)
print("Fifth Model")
fifth_model = Two_NNModel(input_dim=300, hidden_dim=64, output_dim=6)
fifth_model = train_model(model=fifth_model, n_iters=50000, batch_size=50, learning_rate=0.01)
```

```python
# Train best model/validation
# First Model
print("First model")
other_datasets(model=first_model, 
               df_features=validation_df_features, 
               df_labels=validation_df_labels, 
               dataframe_=validation_df)

# Second Model
print("Second model")
other_datasets(model=second_model, 
               df_features=validation_df_features, 
               df_labels=validation_df_labels, 
               dataframe_=validation_df)

# Third Model
print("Third model")
other_datasets(model=third_model, 
               df_features=validation_df_features, 
               df_labels=validation_df_labels, 
               dataframe_=validation_df)

# Fourth Model
print("Fourth model")
other_datasets(model=fourth_model, 
               df_features=validation_df_features, 
               df_labels=validation_df_labels, 
               dataframe_=validation_df)

# Fifth Model
print("Fifth model")
other_datasets(model=fifth_model, 
               df_features=validation_df_features, 
               df_labels=validation_df_labels, 
               dataframe_=validation_df)
```

```python
# Test best model
other_datasets(model=first_model, 
               df_features=test_df_features, 
               df_labels=test_df_labels, 
               dataframe_=test_df)
```
